```python
import pandas as pd
import os
```

```python
#import idx.txt from SEC EDGAR Archives: full-index
idx = pd.read_csv('idx.txt', sep="|",header=None)
idx.columns = ['cik','company','form','fill_date','file']
idx
```

```python
idx['form'].unique()
```

```python
#Keep only mutual fund shareholder report forms
idx1 = idx[(idx['form']=='N-CSR') | (idx['form']=='N-CSRS')]
idx1.drop(columns=['form'], inplace=True)
idx1['file'] = idx1['file'].str[11:-4]
idx1['file'] = idx1['file'].str.replace(r'^\w+/', '', regex=True)
idx1.reset_index(drop=True, inplace=True)
idx1
```

```python
idx1.to_csv('ncsrs_list.csv', index=False)
```
